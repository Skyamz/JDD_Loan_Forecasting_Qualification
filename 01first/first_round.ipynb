```python
import seaborn as sns 

import sklearn
import sklearn.datasets
import sklearn.cross_validation
import sklearn.decomposition
import sklearn.grid_search
import sklearn.neighbors
import sklearn.metrics
```

```python
def accuracy_for_k(k,x,y):
    split_data=sklearn.cross_validation.train_test_split(x,y,test_size=0.33,random_state=99)
    X_train,X_test,Y_train,Y_test=split_data
    knn=sklearn.neighbors.KNeighborsRegressor(n_neighbors=k,weights='uniform')
    knn.fit(X_train,Y_train)
    #Y_hat=knn.predict(X_test)
    value=knn.score(X_test,Y_test)
    return value
```

```python
df = pd.read_csv('/Users/skyamz/Desktop/meanall_fill.csv')
```

```python
test = pd.read_csv('/Users/skyamz/Documents/JDD比赛/Loan_Forecasting_Qualification/t_loan_sum.csv')
```

```python
data = pd.merge(test,df,how='outer',on='uid')
```

```python
del data['loan_sum']
```

```python
X_train = data[:19197]
```

```python
Y_train = test['loan_sum']
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import cross_val_score
```

```python
alphas = np.logspace(-3, 2, 50)
test_scores = []
for alpha in alphas:
    clf = Ridge(alpha)
    test_score = np.sqrt(-cross_val_score(clf, X_train, Y_train, cv=10000, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
plt.plot(alphas, test_scores)
plt.title("Alpha vs CV Error");
```

```python
from sklearn.ensemble import RandomForestRegressor
max_features = [.1, .3, .5, .7, .9, .99]
test_scores = []
for max_feat in max_features:
    clf = RandomForestRegressor(n_estimators=500, max_features=max_feat)
    test_score = np.sqrt(-cross_val_score(clf, X_train, Y_train, cv=20, scoring='neg_mean_squared_error'))
    test_scores.append(np.mean(test_score))
```

```python
plt.plot(max_features, test_scores)
plt.title("Max Features vs CV Error");
```

```python
ridge = Ridge(alpha=100)
rf = RandomForestRegressor(n_estimators=500, max_features=.1)
```

```python
ridge.fit(X_train, Y_train)
rf.fit(X_train, Y_train)
```

```python
# Find the optimal number of k in knn:
k_values=range(1,101)
scores=np.zeros(len(k_values))
for k, c_k in zip(k_values,range(len(k_values))):
    value=accuracy_for_k(k=k,x=X_train,y=Y_train)
    scores[c_k]=value
    
k_opt=np.argmax(scores)+1
print (scores.max())
print ('The optimal value of k is:',k_opt)

sns.tsplot(scores.T)
#plt.xticks(range(len(k_values)),k_values)
plt.xlabel('k',fontsize=18)
plt.ylabel('Accuracy',fontsize=18)
plt.title('Accuracy for different k values.',fontsize=18)
plt.show()
```
