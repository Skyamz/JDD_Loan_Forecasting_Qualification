```python
import pandas as pd
xgb = pd.read_csv('fin_result.csv')
DT = pd.read_csv('DT.csv')
four_result = pd.read_csv('four_result.csv')
SVM_rbf = pd.read_csv('SVM_rbf.csv')
path = "/Users/skyamz/Desktop/JDD_Loan_Forecasting_Qualification/initdata/"
file_user = pd.read_csv(path + "t_user.csv")
```

```python
tmp1 = pd.merge(xgb,DT,how='inner',on='uid')
```

```python
tmp2 = pd.merge(tmp1,four_result,how='inner',on='uid')
```

```python
tmp3 = pd.merge(tmp2,SVM_rbf,how='inner',on='uid')
```

```python
tmp3['loan_sum'] =  0.15*tmp3['xgb']+0.15*tmp3['DT1']+0.2*tmp3['DT2']+0.3*tmp3['four_result']+0.2*tmp3['SVM_rbf']
```

```python
tmp3
```

```python
fifth_result = tmp3[['uid','loan_sum']]
```

```python
fifth = pd.merge(fifth_result,file_user,how='outer',on='uid')
```

```python
fifth = fifth.where(fifth.notnull(),0)
```

```python
fifth.to_csv('fifth_result.csv')
```
